```python
import pandas as pd
import numpy as np
```

```python
#Read in Dataframe
df = pd.read_csv("temperature_and_ghg_concentrations_1880-2014.csv")
```

```python
#Drop index columns
mol = df.drop(["Year", "GCAG", "GISTEMP", "Average Land Temperature (Degrees Celcius)"], axis=1, index=None, columns=None, level=None, inplace=False, errors='raise')
```

```python
#Identify columns to loop through
columns = mol.columns
```

```python
#This is where we change the base year 
#NOTE 89 base year = 1969 always add two/subtract two to reference excel sheet
base_year_1950 = 70

#Do not touch, loops through to find the slope of the line and the first and last values
#We use point slope formula to determine the slope:  m = (y2-y1)/(x2-x1)
m_array_1880to1950 = []
last_value_array_1950 = []
for i in columns:
    r=0
    for row in mol[i]:
        if row == 0:
            r+=1
            next
        else:
            indx1 = r
            first_value = row
            first_point = [indx1,first_value]
            #print(first_point)  
            break
    last_value_1950 = (mol[i].loc[base_year_1950])
    indx2_1950 = base_year_1950
    last_point_1950 = [indx2_1950,last_value_1950]
    last_value_array_1950.append(last_value_1950)
    
    rise_1950 = (last_value_1950 - first_value)
    run_1950 = (indx2_1950 - indx1)
    m_1950 = rise_1950/run_1950
    m_array_1880to1950.append(m_1950)
    next
```

```python
#-------------------Finding slope for 1950 to 1970---------------"THE GREAT ACCELERATION"

#This is where we change the base year 
#NOTE 89 base year = 1969 always add two/subtract two to reference excel sheet
base_year_1970 = 90
first_year_1950 = 70

#Do not touch, loops through to find the slope of the line and the first and last values
#We use point slope formula to determine the slope:  m = (y2-y1)/(x2-x1)
m_array_1950to1970 = []
last_value_array_1970 = []
first_value_array_1950 = []
for i in columns:
    #First Value 1950
    first_value_1950 = (mol[i].loc[first_year_1950])
    indx1_1950 = first_year_1950
    first_point_1950 = [indx1_1950,first_value_1950]
    first_value_array_1950.append(first_value_1950)
    #Last Value 1970
    last_value_1970 = (mol[i].loc[base_year_1970])
    indx2_1970 = base_year_1970
    last_point_1970 = [indx2_1970,last_value_1970]
    last_value_array_1970.append(last_value_1970)
    #Slope for 1950 to 1970
    rise_1950to1970 = (last_value_1970 - first_value_1950)
    run_1950to1970 = (indx2_1970 - indx1_1950)
    m_1950to1970 = rise_1950to1970/run_1950to1970
    m_array_1950to1970.append(m_1950to1970)
    next
```
