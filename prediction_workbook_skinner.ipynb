```python
import pandas as pd
import numpy as np
```

```python
#Read in Dataframe
df = pd.read_csv("temperature_and_ghg_concentrations_1880-2014.csv")
```

```python
#Drop index columns
mol = df.drop(["Year", "GCAG", "GISTEMP", "Average Land Temperature (Degrees Celcius)"], axis=1, index=None, columns=None, level=None, inplace=False, errors='raise')
```

```python
#Identify columns to loop through
columns = mol.columns
```

```python
#This is where we change the base year 
#NOTE 89 base year = 1969 always add two/subtract two to reference excel sheet
base_year_1950 = 70

#Do not touch, loops through to find the slope of the line and the first and last values
#We use point slope formula to determine the slope:  m = (y2-y1)/(x2-x1)
m_array_1880to1950 = []
last_value_array_1950 = []
for i in columns:
    r=0
    for row in mol[i]:
        if row == 0:
            r+=1
            next
        else:
            indx1 = r
            first_value = row
            first_point = [indx1,first_value]
            #print(first_point)  
            break
    last_value_1950 = (mol[i].loc[base_year_1950])
    indx2_1950 = base_year_1950
    last_point_1950 = [indx2_1950,last_value_1950]
    last_value_array_1950.append(last_value_1950)
    
    rise_1950 = (last_value_1950 - first_value)
    run_1950 = (indx2_1950 - indx1)
    m_1950 = rise_1950/run_1950
    m_array_1880to1950.append(m_1950)
    next
```

```python
#-------------------Finding slope for 1950 to 1970---------------"THE GREAT ACCELERATION"

#This is where we change the base year 
#NOTE 89 base year = 1969 always add two/subtract two to reference excel sheet
base_year_1970 = 90
first_year_1950 = 70

#Do not touch, loops through to find the slope of the line and the first and last values
#We use point slope formula to determine the slope:  m = (y2-y1)/(x2-x1)
m_array_1950to1970 = []
last_value_array_1970 = []
first_value_array_1950 = []
for i in columns:
    #First Value 1950
    first_value_1950 = (mol[i].loc[first_year_1950])
    indx1_1950 = first_year_1950
    first_point_1950 = [indx1_1950,first_value_1950]
    first_value_array_1950.append(first_value_1950)
    #Last Value 1970
    last_value_1970 = (mol[i].loc[base_year_1970])
    indx2_1970 = base_year_1970
    last_point_1970 = [indx2_1970,last_value_1970]
    last_value_array_1970.append(last_value_1970)
    #Slope for 1950 to 1970
    rise_1950to1970 = (last_value_1970 - first_value_1950)
    run_1950to1970 = (indx2_1970 - indx1_1950)
    m_1950to1970 = rise_1950to1970/run_1950to1970
    m_array_1950to1970.append(m_1950to1970)
    next
```

```python
#-------------------Finding slope for 1970 to 1988---------------"Enviromental Agencies creation; first reports of global warming"

#This is where we change the base year 
#NOTE 89 base year = 1969 always add two/subtract two to reference excel sheet
base_year_1970 = 90
last_year_1988 = 108

#Do not touch, loops through to find the slope of the line and the first and last values
#We use point slope formula to determine the slope:  m = (y2-y1)/(x2-x1)
m_array_1970to1988 = []
first_value_array_1970 = []
last_value_array_1988 = []
for i in columns:
    #First Value 1970
    first_value_1970 = (mol[i].loc[base_year_1970])
    indx1_1970 = base_year_1970
    first_point_1970 = [indx1_1970,first_value_1970]
    first_value_array_1970.append(first_value_1970)
    #Last Value 1988
    last_value_1988 = (mol[i].loc[last_year_1988])
    indx2_1988 = last_year_1988
    last_point_1988 = [indx2_1988,last_value_1988]
    last_value_array_1988.append(last_value_1988)
    #Slope for 1970 to 1988
    rise_1970to1988 = (last_value_1988 - first_value_1970)
    run_1970to1988 = (indx2_1988 - indx1_1970)
    m_1970to1988 = rise_1970to1988/run_1970to1988
    m_array_1970to1988.append(m_1970to1988)
    next
```

```python
#-------------------Finding slope for 1988to2014 ---------------"Chernobyl in 88; human induced global warming becomes mainstream"

#This is where we change the base year 
#NOTE 89 base year = 1969 always add two/subtract two to reference excel sheet
last_year_2014 = 134
base_year_1988 = 108

#Do not touch, loops through to find the slope of the line and the first and last values
#We use point slope formula to determine the slope:  m = (y2-y1)/(x2-x1)
m_array_1988to2014 = []
first_value_array_1988 = []
last_value_array_2014 = []
for i in columns:
    #First Value 1988
    first_value_1988 = (mol[i].loc[base_year_1988])
    indx1_1988 = base_year_1988
    first_point_1988 = [indx1_1988,first_value_1988]
    first_value_array_1988.append(first_value_1988)
    #Last Value 2014
    last_value_2014 = (mol[i].loc[last_year_2014])
    indx2_2014 = last_year_2014
    last_point_1988 = [indx2_2014,last_value_2014]
    last_value_array_2014.append(last_value_2014)
    #Slope for 1988to2014
    rise_1988to2014 = (last_value_2014 - first_value_1988)
    run_1988to2014 = (indx2_2014 - indx1_1988)
    m_1988to2014 = rise_1988to2014/run_1988to2014
    m_array_1988to2014.append(m_1988to2014)
    next
```
