```python
import pandas as pd
import numpy as np
```

```python
#Read in Dataframe
df = pd.read_csv("temperature_and_ghg_concentrations_1880-2014.csv")
```

```python
#Drop index columns
mol = df.drop(["Year", "GCAG", "GISTEMP", "Average Land Temperature (Degrees Celcius)"], axis=1, index=None, columns=None, level=None, inplace=False, errors='raise')
#mol.head(100)
```

```python
#Identify columns to loop through
columns = mol.columns
```

```python
#This is where we change the base year
base_year = 89

#Do not touch, loops through to find the slope of the line and the first and last values
#We use point slope formula to determine the slope:  m = (y2-y1)/(x2-x1)
m_array = []
last_value_array = []
for i in columns:
    r=0
    for row in mol[i]:
        if row == 0:
            r+=1
            next
        else:
            indx1 = r
            first_value = row
            first_point = [indx1,first_value]
            #print(first_point)  
            break
    last_value = (mol[i].loc[base_year])
    indx2 = base_year
    last_point = [indx2,last_value]
    last_value_array.append(last_value)
    
    rise = (last_value - first_value)
    run = (indx2 - indx1)
    m = rise/run
    m_array.append(m)
    next
```

```python
#This is where we change the interval
year_interval = 1

#This creates our change in concentration array
predict_concentration_change = [i* year_interval for i in m_array]
#predict_concentration_change
```

```python
#Indexing to put in dataframe
target_predict = [last_value_array]
df1 = pd.DataFrame(last_value_array)
df2 = pd.DataFrame(predict_concentration_change)
#adding projected concentration with base year
final_df = df1 + df2
#Set index for merge
predict = final_df.set_index(columns)
```

```python
#Converting back to array
est_concentration = list(final_df[0])
```

```python
actual = pd.DataFrame(mol.iloc[base_year + year_interval])
```

```python
#merged = pd.concat([actual, predict], me)
summary = actual.join(predict, how='outer')
#clean_summary = summary.columns=['Actual','Predict']
clean_summary = summary.rename(columns={90: 'Actual', 0: 'Predict'})

type(clean_summary)
clean_summary
```
